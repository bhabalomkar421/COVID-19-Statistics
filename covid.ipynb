Analysis of cases in India

```python
#reading of Datasets
df = pd.read_csv('Covid_cases_in_India.csv')
df_india = df.copy()
#df_india.head()
```

```python
#Total cases is total no of confirmed cases 

df.drop(['S. No.'],axis=1,inplace=True)
df['Total cases'] = df['Total Confirmed cases (Indian National)'] + df['Total Confirmed cases ( Foreign National )']
total_cases = df['Total cases'].sum()
print("Total number of confirmed cases in india till date (22nd march 2020) is ",total_cases)
```

```python
df.style.background_gradient(cmap='Reds')
```

###Number of Active cases

```python
df['Total Active'] = df['Total cases'] -(df['Death'] + df['Cured'])
total_active = df['Total Active'].sum()
print("Total number of confirmed cases in india till date (22nd march 2020) is  ",total_active)
Tot_cases = df.groupby('Name of State / UT')['Total Active'].sum().sort_values(ascending=False).to_frame()
Tot_cases.style.background_gradient(cmap='Reds')
```

visualising the spread geographically

```python
India_coord = pd.read_csv('Indian Coordinates.csv')
df_full = pd.merge(India_coord,df,on='Name of State / UT')
map = folium.Map(location=[20,70],zoom_start=4,titles='stamenterrain')

for lat,lon,value,name in zip(df_full['Latitude'],df_full['Longitude'],df_full['Total cases'],df_full['Name of State / UT']):
    folium.CircleMarker([lat,lon],radius=value*0.8,popup = ('<strong>State</strong>: ' + str(name).capitalize() + '<br>'))
map
```
